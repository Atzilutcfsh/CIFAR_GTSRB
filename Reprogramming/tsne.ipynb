```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from torchvision.datasets import CIFAR10
```

```python
dataset = CIFAR10(root='.', train=True, download=True)
X = np.array(dataset.data)  # 图像数据
y = np.array(dataset.targets)  # 标签数据
```

```python
X_flat = X.reshape(X.shape[0], -1)
```

```python
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_flat)
```

```python
class_labels = np.unique(y)
class_data = {label: [] for label in class_labels}

for i, label in enumerate(y):
    class_data[label].append(X_tsne[i])
```

```python
plt.figure(figsize=(10, 10))

for label, data in class_data.items():
    data = np.array(data)
    plt.scatter(data[:, 0], data[:, 1], label=str(label), alpha=0.7)

#plt.legend(title='Class')
plt.title('t-SNE Visualization of Dataset')
plt.savefig('tsne_cifar10.pdf', format = 'pdf', dpi = 300)
plt.show()
```
